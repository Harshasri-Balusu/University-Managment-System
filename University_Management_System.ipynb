```python
import random
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText


class Person:
    def __init__(self, rollno, name, email):
        self.name = name
        self.rollno = rollno
        self.email = email


class Student(Person):
    def __init__(self, srollno, sname, semail, branch):
        super().__init__(srollno, sname, semail)
        self.branch = branch


class Teacher(Person):
    def __init__(self, trollno, tname, temail, subject):
        super().__init__(trollno, tname, temail)
        self.subject = subject


class College:
    def __init__(self, cid, cname):
        self.cid = cid
        self.cname = cname
        self.students = []
        self.teachers = []

    def add_student(self, student):
        self.students.append(student)

    def add_teacher(self, teacher):
        self.teachers.append(teacher)


colleges = []


def send_otp(tomail, name, date, month):
    otp = random.randint(1111, 9999)
    months = {
        1: "January", 2: "February", 3: "March", 4: "April",
        5: "May", 6: "June", 7: "July", 8: "August",
        9: "September", 10: "October", 11: "November", 12: "December"
    }
    subject = "OTP For Verification"
    body = f"Hello {name}!\nDate of Birth: {date} - {months[month]}\nYour Secret OTP is {otp}"

    msg = MIMEMultipart()
    msg['From'] = "balusuharshasri@gmail.com"
    msg['To'] = tomail
    msg['Subject'] = subject
    msg.attach(MIMEText(body, 'plain'))


    server = smtplib.SMTP("smtp.gmail.com", 587)
    server.starttls()
    server.login("balusuharshasri@gmail.com", "duii lebb gpjj egvc")
    server.send_message(msg)
    server.quit()

    return otp


def verify_otp():
    name = input("Enter Your Name: ")
    date = int(input("Enter Your Date of Birth (Day): "))
    month = int(input("Enter Your Month of Birth (1-12): "))
    tomail = input("Enter Your Email ID: ")
    otp = send_otp(tomail, name, date, month)

    user_input = input("Enter the OTP sent to your email: ")
    if user_input == str(otp):
        print("OTP Validation Successful!")
        return True
    else:
        print("Wrong OTP Entered!")
        return False


while True:
    print("\nChoose Your Option:")
    print("1. Add College")
    print("2. Add Student")
    print("3. Add Teacher")
    print("4. Display Student Details")
    print("5. Display Teacher Details")
    print("6. Display Students of a Particular Branch")
    print("7. Display Teachers of a Particular Subject")
    print("8. Exit")
    ip = int(input("Enter Your Option: "))

    if ip == 1:
        cname = input("Enter College Name: ")
        cid = input("Enter College ID: ")
        if any(c.cid == cid for c in colleges):
            print("College Already Exists!")
        else:
            colleges.append(College(cid, cname))
            print("College Created Successfully!")
    elif ip == 2:
        cid = input("Enter College ID: ")
        clg = next((c for c in colleges if c.cid == cid), None)
        if clg and verify_otp():
            name = input("Enter Student Name: ")
            roll = input("Enter Student Roll Number: ")
            email = input("Enter Student Email: ")
            branch = input("Enter Student Branch: ")
            clg.add_student(Student(roll, name, email, branch))
            print("Student Added Successfully!")
        else:
            print("Invalid College ID or OTP Verification Failed.")
    elif ip == 3:
        cid = input("Enter College ID: ")
        clg = next((c for c in colleges if c.cid == cid), None)
        if clg and verify_otp():
            name = input("Enter Teacher Name: ")
            roll = input("Enter Teacher Roll Number: ")
            email = input("Enter Teacher Email: ")
            subject = input("Enter Teacher Subject: ")
            clg.add_teacher(Teacher(roll, name, email, subject))
            print("Teacher Added Successfully!")
        else:
            print("Invalid College ID or OTP Verification Failed.")
    elif ip == 4:
        cid = input("Enter College ID: ")
        clg = next((c for c in colleges if c.cid == cid), None)
        if clg:
            print(f"Student Details of {clg.cname}:")
            for s in clg.students:
                print(f"Roll No: {s.rollno}, Name: {s.name}, Email: {s.email}, Branch: {s.branch}")
        else:
            print("Invalid College ID.")
    elif ip == 5:
        cid = input("Enter College ID: ")
        clg = next((c for c in colleges if c.cid == cid), None)
        if clg:
            print(f"Teacher Details of {clg.cname}:")
            for t in clg.teachers:
                print(f"Roll No: {t.rollno}, Name: {t.name}, Email: {t.email}, Subject: {t.subject}")
        else:
            print("Invalid College ID.")
    elif ip == 6:
        branch = input("Enter the Branch: ")
        for clg in colleges:
            print(f"Students of Branch {branch} in {clg.cname}:")
            for s in clg.students:
                if s.branch == branch:
                    print(f"Roll No: {s.rollno}, Name: {s.name}, Email: {s.email}")
    elif ip == 7:
        subject = input("Enter the Subject: ")
        for clg in colleges:
            print(f"Teachers of Subject {subject} in {clg.cname}:")
            for t in clg.teachers:
                if t.subject == subject:
                    print(f"Roll No: {t.rollno}, Name: {t.name}, Email: {t.email}")
    elif ip == 8:
        print("Thank You! Exiting...")
        break
    else:
        print("Invalid Option! Please Try Again.")
```
